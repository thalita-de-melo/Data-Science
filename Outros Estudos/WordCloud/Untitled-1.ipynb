```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# Start with one review:
text = 'Educao e Sade so as reas mais afetadas, mas antes mesmo do anncio do bloqueio j havia setores prejudicados por falta de verbas, como o de emisso de passaportes e de manuteno de carros da Polcia Rodoviria Federal. O objetivo do novo contingenciamento  cumprir a regra do teto de gastos. Esta norma, prevista na legislao, estabelece que a maior parte das despesas no pode subir acima da inflao do ano anterior. Como as despesas obrigatrias (salrios de servidores, gastos previdencirios e com assistncia social, entre outras) tm subido a cada ano, o espao para os chamados gastos discricionrios, que no so obrigatrios, tm diminudo.'
stopwords = ['ver','principal','essa','vez','nas','mas','qual','principal','ele','ter','doena','pois','este','vez','ver principal','artigo principal','j','aos','pode','outro','artigo','desse','alguns','meio','entre','das','podem','esse','seu','tambm','so','quando','de', 'que','em','os','as','da','como','dos','ou','se','um','uma','para','na','ao','mais','por','no','ainda','muito','sua'] + list(STOPWORDS)

# Create and generate a word cloud image:
wordcloud = WordCloud(stopwords=stopwords).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
